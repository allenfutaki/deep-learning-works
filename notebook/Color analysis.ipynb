```python
import numpy as np
import cv2

bright = cv2.imread('/home/allen/Downloads/bright.jpg')
dark = cv2.imread('/home/allen/Downloads/dark.jpg')

brightLAB = cv2.cvtColor(bright, cv2.COLOR_BGR2LAB)
darkLAB = cv2.cvtColor(dark, cv2.COLOR_BGR2LAB)

#python
bgr = [36,24,20]
thresh = 40
 
#convert 1D array to 3D, then convert it to LAB and take the first element 
lab = cv2.cvtColor( np.uint8([[bgr]] ), cv2.COLOR_BGR2LAB)[0][0]
 
minLAB = np.array([lab[0] - thresh, lab[1] - thresh, lab[2] - thresh])
maxLAB = np.array([lab[0] + thresh, lab[1] + thresh, lab[2] + thresh])
 
maskLAB = cv2.inRange(brightLAB, minLAB, maxLAB)
resultLAB = cv2.bitwise_and(brightLAB, brightLAB, mask = maskLAB)

cv2.imwrite("/home/allen/Downloads/result.jpg", resultLAB)
```

```python
import pandas as pd
import re
import os.path as osp
from PIL import Image
def to_pil(cv_img):
    img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB)
    return Image.fromarray(img)
```

```python
df = pd.read_csv("/media/allen/mass/office_color/office_color.csv")

df = df.drop('Unnamed: 0', axis=1)

src = "/media/allen/mass/office/office/"

df.img = df.img.apply(lambda x: osp.join(src, x))    

# Color = LAB(accurate=True)

pattern = re.compile(r'\[(\d+), (\d+), (\d+), (\d+)\]')
```
