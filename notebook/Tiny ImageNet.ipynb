```python
index_map = {}
with open('/media/allen/mass/tiny-imagenet-200/wnids.txt') as f:
    for i, line in enumerate(f.readlines()):
        line = line.strip().split(" ")
        index_map[line[0]] = i
```

```python
dataset = []
with open('/media/allen/mass/tiny-imagenet-200/val/val_annotations.txt') as f:
    for i, line in enumerate(f.readlines()):
        line = line.strip().split("\t")
        dataset.append(line[:2])
with open('/media/allen/mass/DB/tinyimagenet/tiny_val.txt', 'w') as f:
    for img_name, cls_text in dataset:
        f.write(f"{img_name} {index_map[cls_text]}\n")
```
