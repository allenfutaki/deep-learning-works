```python
import h5py
model_path = '/home/allen/fmobilenet.h5'
f = h5py.File(model_path, 'r')
```

```python
for key, value in f.attrs.items():           
    if key == 'layer_names':
        for layer in f.attrs[key]:
            for group in f[layer].keys():
                for weight_name in f[layer][group].keys():
                    print("{:30}-{:20} ---> {}".format(layer, weight_name, f[layer][group][weight_name][:].shape))
```
