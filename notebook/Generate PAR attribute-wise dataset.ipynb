```python
src_dir = '/media/allen/mass/PAR'
dataset = []
with open('/media/allen/mass/PAR/PAR_train_dataset.txt', 'r') as f:
    count = 0
    pack = 0
    for line in f:
        line = line.strip().split(' ')
        path = line[0]
        extension = path.split(".")[-1]
        x1, y1, x2, y2 = int(line[1]), int(line[2]), int(line[3]), int(line[4])
        if x2 < x1:            
            x2 += x1
            y2 += y1
        
        s = '_'
        attr = s.join(line[5:])
        img = cv2.imread(osp.join(src_dir, path))
        if x1 > -1:
            crop = img[y1:y2, x1:x2, :]
        else:
            crop = img
        
        if not os.path.exists("/media/allen/mass/PAR_ATTR/train/{}".format(pack)):
            os.mkdir("/media/allen/mass/PAR_ATTR/train/{}".format(pack))
                     
        dst = "/media/allen/mass/PAR_ATTR/train/{}/{:06}__{}.{}".format(pack, count, attr, extension)
                     
        cv2.imwrite(dst, crop)
        count += 1
        if count % 10000 == 0:
            pack += 1
        
```
