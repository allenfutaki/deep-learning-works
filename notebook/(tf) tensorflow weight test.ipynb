```python
def Conv_layer(x):
    conv_output = tf.layers.conv2d(
                    inputs=x,
                    filters=1,
                    kernel_size=(3,3),
                    strides=(2,2),
                    padding='valid',
                    data_format='channels_last',
                    dilation_rate=(1, 1),
                    activation=None,
                    use_bias=False,
                    kernel_initializer=tf.ones_initializer()
                )
    return conv_output
```

```python
def dw_layer(inputs):
    np.random.seed(5)
    npfilter1 = np.random.rand(4,4,1,1).tolist()
    npfilter2 = np.random.rand(3,3,1,1).tolist()
    npfilter3 = np.random.rand(3,3,1,1).tolist()
    filter1 = tf.constant(value=npfilter1, dtype=tf.float32)
    filter2 = tf.constant(value=npfilter2, dtype=tf.float32)
    filter3 = tf.constant(value=npfilter3, dtype=tf.float32)
    kernel = tf.compat.v1.concat(values=[filter1,filter2,filter3],axis=2)
    output = tf.nn.depthwise_conv2d(inputs, filter=kernel, strides=[1,1,1,1], rate=[1,1], padding='VALID')
    return output
```
